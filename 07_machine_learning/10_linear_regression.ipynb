# Linear Regression

Linear regression is a statistical method used to model the relationship between a **dependent variable (target)** and one or more **independent variables (predictors)**. The goal of linear regression is to find the best-fitting line (or hyperplane in higher dimensions) that minimizes the difference between the predicted and actual values.

### Equation of Linear Regression:
\[
y = \beta_0 + \beta_1x + \epsilon
\]

Where:
- \( y \): Dependent variable (target)
- \( x \): Independent variable (predictor)
- \( \beta_0 \): Intercept of the line
- \( \beta_1 \): Coefficient (slope) of the independent variable
- \( \epsilon \): Error term (residual)

### Key Features:
- **Simple Linear Regression**: Involves one independent variable.
- **Multiple Linear Regression**: Involves two or more independent variables.
- It assumes a linear relationship between the predictors and the target variable.
- Used for tasks such as prediction and understanding the strength and nature of relationships between variables.


Simple Linear Regression

```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# load the data tips from sns
df = sns.load_dataset('tips')
df.head()
```

```python
sns.scatterplot(x='total_bill', y='tip', data=df)
```

```python
# split the data into X and y
X = df[['total_bill']]
# scalar = MinMaxScaler()
# X = scalar.fit_transform(X)
y = df['tip']
```

```python
# split the data into train and test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
# call the model
model = LinearRegression()
```

```python
# train the model
model.fit(X_train, y_train)
```

```python
# take out model intercept and slop, make an equation
print(model.intercept_)
print(model.coef_)
print('y = ', model.intercept_, '+', model.coef_, '* X')
```

```python
model.predict([[5]])
```

```python
# predict
y_pred = model.predict(X_test)
```

```python
# evaluate the model
print('MSE = ', mean_squared_error(y_test, y_pred))
print('R2 = ', r2_score(y_test, y_pred))
print('RMSE = ', np.sqrt(mean_squared_error(y_test, y_pred)))
```

```python
# plot the model and data
plt.scatter(X_test, y_test)
plt.plot(X_test, y_pred, color='red')
plt.show()
```
