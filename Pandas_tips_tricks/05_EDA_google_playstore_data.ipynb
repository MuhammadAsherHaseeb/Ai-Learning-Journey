```python
# Convert the whole Size column into bytes
# lets define a function
def convert_size(size):
    if isinstance(size, str):
        if 'k' in size: # shift + 9 to add bracket to string
            return float(size.replace('k', "")) *1024
        elif 'M' in size:
            return float(size.replace('M',"")) * 1024 * 1024
        elif 'Varies with device' in size:
            return np.nan
    return size
```

```python
# lets apply this function
df["Size"] = df['Size'].apply(convert_size)
```

```python
# check
df['Size']
```

```python
# rename the column
df.rename(columns={'Size': "Size_in_bytes"}, inplace=True)
```

Conver t Size in MBS

```python
df['Size_in_Mb'] = df['Size_in_bytes'].apply(lambda x: x/(1024*1024))
```

Conver Size in Kb

```python
df['Size_in_kb'] = df['Size_in_bytes'].apply(lambda x: x/(1024))
```

### Observations on Installs
1. Remove + Sign
2. Remove ,
3. Convert it to Integer

```python
# reobjectifying we need to write inplace true again
df['Installs'] = df['Installs'].apply(lambda x: x.replace('+',"") if '+' in str(x) else x)
```

```python
# removing , from all columns
df['Installs'] = df['Installs'].apply(lambda x: x.replace(',',"") if ',' in str(x) else x)
```

```python
# covert it to integer using lambda function
df['Installs'] = df['Installs'].apply(lambda x:int(x))
```

# Price Column

```python
df['Price'].value_counts()
```

### Observations in Price
1. $

```python
# lets see how many values have $ sign
df['Price'].loc[df['Price'].str.contains('\$')].value_counts().sum()
```

```python
# and who dont have $ sign
df['Price'].loc[(df['Price'].str.contains('0')) & (~df['Price'].str.contains('\$'))].value_counts().sum()
```

```python
# removing $ from price Price Column
df['Price'] = df['Price'].apply(lambda x: x.replace('$',"") if '$' in str(x) else x)
```

```python
df['Price'].value_counts
```

```python
# now we can covert it to float column
# because Price are continous 
df['Price'] = df['Price'].apply(lambda x:float(x))
```

```python
# plot missing values
plt.figure(figsize=(16,6))
sns.heatmap(df.isnull(), yticklabels=False, cbar=False, cmap='viridis')
```

Plot of misiing values in percentage

```python
# make figure size

plt.figure(figsize=(16, 6))
# plot the null values by their percentage in each column
missing_percentage = df.isnull().sum()/len(df)*100
missing_percentage.plot(kind='bar')
# add the labels
plt.xlabel('Columns')
plt.ylabel('Percentage')
plt.title('Percentage of Missing Values in each Column')
```

```python
# plot missing values
missing_percentage[missing_percentage < 1].plot(kind='bar')
```
