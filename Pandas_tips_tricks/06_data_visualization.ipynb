```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# import titanic data set suing seaborn library
df = sns.load_dataset('titanic')
```

```python
df.plot(kind='scatter', x='age', y='fare')  
```

```python
df.plot(kind='scatter', x='age', y='fare', color = 'red')
```

```python
# changing color using hex codes
df.plot(kind='scatter', x='age', y='fare', color = '#FE7F0E')  
```

```python
#grouping
df.groupby('sex')['age'].mean()
```

```python
#grouping and ploting
df.groupby('sex')['fare'].mean().plot(kind='bar')
# labling 
plt.ylabel('Average Fare')
plt.xlabel('Gender')
```

```python
# Using plt to create plots
plt.plot
```

```python
# seaborn plots
# bar plot using seasborn on df
sns.barplot(data = df, x = 'survived', y = 'fare')
```

```python
# seaborn plots
# bar plot using seasborn on df and coloring
sns.barplot(data = df, x = 'survived', y = 'fare', hue = 'class')
```

```python
# Boxplot using same code
sns.boxplot(data = df, x = 'sex', y = 'age', hue = 'class' )
```

```python
# Flipping the graph
sns.scatterplot(data = df, x= 'fare', y = 'age', hue = 'class')
```

```python
 sns.scatterplot(data = df, y = 'fare', x = 'age', hue = 'class')
```

```python
# creating interactive plot
import plotly.express as px
fig = px.scatter(df, x= 'age', y = 'fare', color = 'class')
fig.show()
```

```python
# creating  3D plot
import plotly.graph_objects as go

import pandas as pd

# Read data from a csv
z_data = pd.read_csv('https://raw.githubusercontent.com/plotly/datasets/master/api_docs/mt_bruno_elevation.csv')

fig = go.Figure(data=[go.Surface(z=z_data.values)])

fig.update_layout(title=dict(text='Mt Bruno Elevation'), autosize=False,
                  width=500, height=500,
                  margin=dict(l=65, r=50, b=65, t=90))

fig.show()
```

```python
# animated plot with plotly
import plotly.express as px
df = px.data.gapminder()
px.scatter(df, x="gdpPercap", y="lifeExp", animation_frame="year", animation_group="country",
           size="pop", color="continent", hover_name="country",
           log_x=True, size_max=55, range_x=[100,100000], range_y=[25,90])
```
