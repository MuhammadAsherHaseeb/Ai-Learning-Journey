```python
# Importing dataset 
df = sns.load_dataset('titanic')
```

```python
df.info()
```

```python
# Scatter Plot
# size of the figure
plt.figure(figsize = (10, 7)) # A4 Size page Ration
sns.scatterplot(data = df, x= 'age', y = 'fare', hue = 'class')
```

```python
# Scatter Plot
# size of the figure
plt.figure(figsize = (10, 7)) # A4 Size page Ration
# sns.scatterplot(data = df, x= 'age', y = 'fare', hue = 'sex')
# changing the scatter point colors of the hue
palette = {'male': 'green', 'female':'red'}
sns.scatterplot(data=df, x='age',y='fare', hue='sex', palette=palette)
# LAbels
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Scatter Plot of Age vs Fare')
# Saving a Plot a plot  must be 300dpi, 10dpi and rare 600dpi in png
plt.savefig('scatter_plot.png', dpi=300) 
# saving in svf
# plt.savefig('scatter_plot.svg')
```

```python
# style of the plot
sns.set_style('darkgrid')
# Scatter Plot
# size of the figure
plt.figure(figsize = (10, 7)) # A4 Size page Ration
# sns.scatterplot(data = df, x= 'age', y = 'fare', hue = 'sex')
# changing the scatter point colors of the hue
palette = {'male': 'green', 'female':'red'}
sns.scatterplot(data=df, x='age',y='fare', hue='sex', palette=palette)
# LAbels
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Scatter Plot of Age vs Fare')
# Saving a Plot a plot  must be 300dpi, 10dpi and rare 600dpi in png
plt.savefig('scatter_plot.png', dpi=300) 
```

# Box Plot

```python
# Box plot
# style of the plot
sns.set_theme(style='darkgrid', font_scale=1.5, font='serif') #use font and font scale
# Scatter Plot
# size of the figure
plt.figure(figsize = (8,4)) # A4 Size 16:9
# sns.scatterplot(data = df, x= 'age', y = 'fare', hue = 'sex')
# changing the scatter point colors of the hue
palette = {'male': 'green', 'female':'red'}
sns.boxplot(data=df, x='sex',y='fare', hue='sex', palette=palette)
# LAbels
plt.xlabel('Age')
plt.ylabel('Fare') 
```

```python
df['class'].unique()
```

```python
# Box plot
# style of the plot
sns.set_theme(style='darkgrid', font_scale=1, font='Times New Roman') #use font and font scale
# Scatter Plot
# size of the figure
plt.figure(figsize = (12,6)) # A4 Size 16:9
# sns.scatterplot(data = df, x= 'age', y = 'fare', hue = 'sex')
# changing the scatter point colors of the hue
palette = {'First': 'green', 'Second':'red', 'Third':'blue'}
sns.boxplot(data=df, x='sex',y='age', hue='class', palette=palette)
# LAbels
plt.xlabel('Age')
plt.ylabel('Fare') 
plt.title('Box Plot of Fare vs Class')

# change legend position and titile
plt.legend(loc='upper right', title='Class Name')
```

```python
# Convert it to a Bubble Plot
# # Box plot
# style of the plot
sns.set_theme(style='darkgrid', font_scale=1.5, font='serif') #use font and font scale
# Scatter Plot
# size of the figure
plt.figure(figsize = (16,6)) # A4 Size 16:9
palette = {'male': 'green', 'female':'red'}
sns.scatterplot(data=df, x='age',y='fare', hue='sex', size ='fare', palette=palette)
# LAbels
plt.xlabel('Age')
plt.ylabel('Fare') 
plt.title('scatter Plot of Age vs Fare') 
```

```python
# Boxen plot
# style of the plot
sns.set_theme(style='darkgrid', font_scale=1, font='Times New Roman') #use font and font scale
# Scatter Plot
# size of the figure
plt.figure(figsize = (12,6)) # A4 Size 16:9
# sns.scatterplot(data = df, x= 'age', y = 'fare', hue = 'sex')
# changing the scatter point colors of the hue
palette = {'First': 'green', 'Second':'red', 'Third':'blue'}
sns.boxenplot(data=df, x='sex',y='age', hue='class', palette=palette)
# LAbels
plt.xlabel('GENDER')
plt.ylabel('Age') 
plt.title('Box Plot of Age based on Class of Passengers')

# change legend position and titile
plt.legend(loc='upper right', title='Class Name')
```

```python
# Boxen plot
# style of the plot
sns.set_theme(style='darkgrid', font_scale=1, font='Times New Roman') #use font and font scale
# Scatter Plot
# size of the figure
plt.figure(figsize = (12,6)) # A4 Size 16:9
# sns.scatterplot(data = df, x= 'age', y = 'fare', hue = 'sex')
# changing the scatter point colors of the hue
palette = {'First': 'green', 'Second':'red', 'Third':'blue'}
sns.boxenplot(data=df, x='sex',y='age', hue='class', palette=palette)
# LAbels
plt.xlabel('GENDER')
plt.ylabel('Age') 
plt.title('Box Plot of Age based on Class of Passengers')

# change legend position and titile
plt.legend(loc='upper right', title='Class Name')

# change the x tick labels
plt.xticks(ticks=[0,1], label=["Male", "Female"],)
# rotate the x tick labels
plt.xticks(rotation=45)
```

```python
# make scatter plot with plotly
import plotly.express as px

fig = px.scatter(df, x= 'age', y = 'fare', color = 'class')
fig.show()
```

```python
# BAr Plot using Seacorn
# Boxen plot
# style of the plot
sns.set_theme(style='darkgrid', font_scale=1, font='Times New Roman') #use font and font scale
# Scatter Plot
# size of the figure
plt.figure(figsize = (12,6)) # A4 Size 16:9
# sns.scatterplot(data = df, x= 'age', y = 'fare', hue = 'sex')
# changing the scatter point colors of the hue
palette = {'First': 'green', 'Second':'red', 'Third':'blue'}
sns.barplot(data=df, x='sex',y='age', hue='class', palette=palette, capsize =0.1) # add caps to the error bars
# LAbels
plt.xlabel('GENDER')
plt.ylabel('Age') 
plt.title('Bar Plot of Age based on Class of Passengers')

# change legend position and titile
plt.legend(loc='upper right', title='Class Name')

# change the x tick labels
plt.xticks(ticks=[0,1], label=["Male", "Female"],)
# rotate the x tick labels
plt.xticks(rotation=45)
# add caps to the error bars
# plt.errorbar(x=df['age'], y=df['fare'], yerr=df['fare'], fmt='o', capsize=5)
```

```python
# Violoin Plot using Seacorn
# Boxen plot
# style of the plot
sns.set_theme(style='darkgrid', font_scale=1, font='Times New Roman') #use font and font scale
# Scatter Plot
# size of the figure
plt.figure(figsize = (12,6)) # A4 Size 16:9
# sns.scatterplot(data = df, x= 'age', y = 'fare', hue = 'sex')
# changing the scatter point colors of the hue
palette = {'First': 'green', 'Second':'red', 'Third':'blue'}
sns.violinplot(data=df, x='sex',y='age', hue='class', palette=palette,) # no caps to the error bars
# LAbels
plt.xlabel('GENDER')
plt.ylabel('Age') 
plt.title('Bar Plot of Age based on Class of Passengers')

# change legend position and titile
plt.legend(loc='upper right', title='Class Name')

# change the x tick labels
plt.xticks(ticks=[0,1], label=["Male", "Female"],)
# rotate the x tick labels
plt.xticks(rotation=45)
# add caps to the error bars
# plt.errorbar(x=df['age'], y=df['fare'], yerr=df['fare'], fmt='o', capsize=5)
```
